### Zadanie domowe - binaryzacja adaptacyjna w oknach z interpolacją.

Pokazana w ramach podstawowej części ćwiczenia binaryzacja adaptacyjna działa dobrze, ale jest dość złożona obliczeniowo (choć oczywiście należy mieć świadomość, że implementację można zoptymalizować i wyeliminować pewne powtarzające się obliczenia).
Zbliżone rozwiązanie można również realizować w nieco innym wariancie - w oknach.
Ogólna idea jest następująca: wejściowy obraz dzielimy na nienachodzące (rozłączne) okna - wygodnie jest założyć, że są one kwadratowe i o rozmiarze będącym potęgą liczby 2.
W każdym z okien obliczamy próg - niech to będzie średnia i stosujemy do binaryzacji lokalnej.
Jak nietrudno się domyślić efekt nie będzie dobry, gdyż na granicach okien wystąpią artefakty.
Aby je wyeliminować należy zastosować interpolację, co zostanie szczegółowo opisane poniżej.
Warto zaznaczyć, że podobny mechanizm interpolacji stosowany jest w poznanym wcześniej algorytmie CLAHE.
Zainteresowane osoby odsyłam do artykułu na [Wikipedii](https://en.wikipedia.org/wiki/Adaptive_histogram_equalization) oraz do artykułu o metodzie CLAHE - Zuiderveld, Karel. “Contrast Limited Adaptive Histograph Equalization.” Graphic Gems IV. San Diego: Academic Press Professional, 1994. 474–485.

```python
rice_img = cv2.imread('rice.png', cv2.IMREAD_GRAYSCALE)
(X, Y) = rice_img.shape

plt.imshow(rice_img, 'gray')
plt.axis('off')
plt.show()
```

```python
def window_binarization(image, W=16):
    averages = np.zeros((X//W, Y//W))
    thresholds = np.zeros_like(image, dtype=np.uint8)

    for i in range(X//W):
        for j in range(Y//W):
            averages[i, j] = np.mean(image[i*W:i*W+W, j*W:j*W+W])

    for i in range(X):
        for j in range(Y):
            thresholds[i, j] = averages[i // W, j // W]

    return thresholds
```

```python
def plot_thresholds(image, thresholds):
    fig, axs = plt.subplots(1, 3, figsize=(15, 15))

    axs[0].set_title('Original image')
    axs[0].imshow(image, 'gray')
    axs[0].axis('off')

    axs[1].set_title('Thresholds')
    axs[1].imshow(thresholds, 'gray')
    axs[1].axis('off')

    axs[2].set_title('Binarized image')
    axs[2].imshow((image > thresholds).astype(np.uint8), 'gray')
    axs[2].axis('off')

    plt.show()
```

```python
rice_thresholds = window_binarization(rice_img, W=16)
```

```python
plot_thresholds(rice_img, rice_thresholds)
```

```python
def binarization_with_interpolation(image, W=16):
    (X, Y) = image.shape

    averages = np.zeros((X//W, Y//W))
    thresholds = np.zeros_like(image, dtype=np.uint8)
    for i in range(Y//W):
        for j in range(X//W):
            averages[i, j] = np.mean(image[i*W:i*W + W, j*W:j*W + W])

    for i in range(Y):
        for j in range(X):
            row_index = i // W
            column_index = j // W

            threshold = None

            # corner
            if (i < W/2 or i >= X - W/2) and (j < W/2 or j >= Y - W/2):
                threshold = averages[row_index, column_index]

            # vertical
            elif j < W/2 or j >= X - W/2:
                row_index = int(i - W/2) // W

                t1 = averages[row_index, column_index]
                t2 = averages[row_index + 1, column_index]

                dY1 = i - W/2 - row_index*W
                dY2 = -i + W/2 + (row_index + 1)*W

                threshold = (dY1 * t2 + dY2 * t1) // W

            # horizontal
            elif i < W / 2 or i >= Y - W / 2:
                column_index = int(j - W / 2) // W

                t1 = averages[row_index, column_index]
                t2 = averages[row_index, column_index + 1]

                dX1 = j - W / 2 - column_index*W
                dX2 = -j + W / 2 + (column_index + 1)*W

                threshold = (dX1 * t2 + dX2 * t1) // W

            # center
            else:
                row_index = int(i - W/2)//W
                column_index = int(j - W/2)//W

                dX1 = j - W/2 - column_index*W
                dX2 = -j + W/2 + (column_index + 1)*W 
                dY1 = i - W/2 - row_index*W
                dY2 = -i + W/2 + (row_index + 1)*W 

                t11 = averages[row_index, column_index]
                t12 = averages[row_index, column_index + 1]
                t21 = averages[row_index + 1, column_index]
                t22 = averages[row_index + 1, column_index + 1]

                t1 = (dX2 * t11 + dX1 * t12) / W
                t2 = (dX2 * t21 + dX1 * t22) / W

                threshold = (dY2 * t1 + dY1 * t2) / W
            
            thresholds[i, j] = threshold

    return thresholds
```

```python
rice_inter_thresholds_16 = binarization_with_interpolation(rice_img, W=16)
plot_thresholds(rice_img, rice_inter_thresholds_16)
```

```python
rice_inter_thresholds_32 = binarization_with_interpolation(rice_img, W=32)
plot_thresholds(rice_img, rice_inter_thresholds_32)
```

```python
rice_inter_thresholds_64 = binarization_with_interpolation(rice_img, W=64)
plot_thresholds(rice_img, rice_inter_thresholds_64)
```

```python
def otsu(img):
    threshold, _ = cv2.threshold(img, 127, 256, cv2.THRESH_OTSU)
    thresholds = np.ones_like(img, dtype=np.uint8)

    thresholds *= int(threshold)
    return thresholds
```

```python
def local_binarization(img, W=15):
    (X, Y) = img.shape

    thresholds = np.zeros_like(img, dtype=np.uint8)
    img_padded = np.pad(img, pad_width=W//2, constant_values=-1)

    for j in range(Y):
        for i in range(X):
            window = img_padded[i:i+W, j:j + W]
            thresholds[i, j] = np.mean(window[window >= 0])

    return thresholds
```

```python
def local_binarization_sauvol_pietikainen(img, W=15, R=128, k=0.15):
    (X, Y) = img.shape

    thresholds = np.zeros_like(img, dtype=np.uint8)
    img_padded = np.pad(img, pad_width=W//2, constant_values=-1)

    for j in range(Y):
        for i in range(X):
            window = img_padded[i:i+W, j:j + W]
            thresholds[i, j] = np.mean(window[window >= 0])*(1 + k*(np.std(window[window >= 0])/R - 1))

    return thresholds
```

```python
def compare_binarizations(img, savouli_k=0.15, W_windowed=(16, 64)):
    titles = [
        "Otsu",
        "Local Average, W=15",
        "Local Sauvoli, W=15",
        f"Windowed W={W_windowed[0]}",
        f"Windowed with interpolation W={W_windowed[1]}"
    ]

    thresholds = [
        otsu(img),
        local_binarization(img),
        local_binarization_sauvol_pietikainen(img, k=savouli_k),
        window_binarization(img, W=W_windowed[0]),
        binarization_with_interpolation(img, W=W_windowed[1])
    ]

    fig, axs = plt.subplots(3, 5, figsize=(15, 10))
    for i in range(5):
        axs[0, i].set_title(titles[i])

        axs[0, i].imshow(img, 'grey')
        axs[0, i].axis('off')

        axs[1, i].imshow(thresholds[i], 'grey')
        axs[1, i].axis('off')

        axs[2, i].imshow((img > thresholds[i]).astype(np.uint8), 'grey')
        axs[2, i].axis('off')

    fig.tight_layout()
    plt.show()
```

```python
compare_binarizations(rice_img, savouli_k=-0.15)
```
